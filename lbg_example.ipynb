# Lined-Buzo-Gray algorithm for clustering

## Example 1
Run LBG on randomly generated dataset

```python
# create a random data set
import numpy as np
np.random.seed(0)
training_data = np.random.rand(100, 2)
```

### size of code book = 4

```python
# import function and run on training data
import lbg
codebook, abs_weights, rel_weights = \
    lbg.generate_codebook(data=training_data, size_codebook=4, epsilon=0.00001, verbose=True)
```

```python
# visualize results
import matplotlib.pyplot as plt

f, ax = plt.subplots()

ax.scatter(training_data[:, 0], training_data[:, 1], label='Training Data', s=5, alpha=0.5)

for codevec_idx, codevec in enumerate(codebook):
    ax.scatter(codevec[0], codevec[1], label=f'Code Vector {codevec_idx+1}', s=60)

# adjust legend
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_title('Clustering by Lined-Buzo-Gray algorithm')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```

## Example 2
Run LBG on synthetic data with clusters.

```python
# create data clusters
cluster_1 = np.random.multivariate_normal(mean=[0, 0], cov=([[1, 0], [0, 1]]), size=10)
cluster_2 = np.random.multivariate_normal(mean=[2, 4], cov=([[3, 1], [1, 3]]), size=16)
cluster_3 = np.random.multivariate_normal(mean=[-1, 0.5], cov=([[0.5, 0.2], [0.2, 0.45]]), size=20)
cluster_4 = np.random.multivariate_normal(mean=[0.5, -2], cov=([[1.5, 0.8], [0.8, 1]]), size=20)

# concatenate all data points into training data
training_data = np.concatenate((cluster_1, cluster_2, cluster_3, cluster_4), axis=0)
rng = np.random.default_rng(seed=1)
rng.shuffle(training_data)
training_data
```

### size of code book = 4

```python
# cluster using LBG
import lbg
codebook, abs_weights, rel_weights = \
    lbg.generate_codebook(data=training_data, size_codebook=4, epsilon=0.00001, verbose=True)
```

```python
# visualize results
import matplotlib.pyplot as plt

f, ax = plt.subplots()

ax.scatter(training_data[:, 0], training_data[:, 1], label='Training Data', s=5, alpha=0.5)

for codevec_idx, codevec in enumerate(codebook):
    ax.scatter(codevec[0], codevec[1], label=f'Code Vector {codevec_idx+1}', s=60)

# adjust legend
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.set_title('Clustering by Lined-Buzo-Gray algorithm')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
```
